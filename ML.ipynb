```python
# Import Library
import pandas
# Import Data
data = pandas.read_csv("data.csv")
data
```

```python
# Klasifikasi pengalaman Kerja Function
def getClassifyPengalamanKerja(data_PengalamanKerja):
    dataClassifyPengalamanKerja = []
    Diterima = 0
    Dipertimbangkan = 0
    Ditolak = 0
    i = 0
    for pengalaman in data_PengalamanKerja:
        if pengalaman <= 10:
            Diterima += 1
            dataClassifyPengalamanKerja.append("Junior")
            i+=1
        if pengalaman > 10 and pengalaman <= 20:
            Dipertimbangkan += 1
            dataClassifyPengalamanKerja.append("Supervisor")
            i+=1
        if pengalaman >= 21 and pengalaman <= 30:
            Ditolak += 1
            dataClassifyPengalamanKerja.append("Senior")
            i+=1
    return dataClassifyPengalamanKerja
```

```python
# Klasifikasi Usia Function
def getClassifyUsia(data_usia):
    dataClassifyUsia =  []
    Tua = 0
    Parubaya = 0
    Muda = 0
    i = 0
    for usia in data_usia:
        if usia >= 46 and usia <= 55:
            Tua += 1
            dataClassifyUsia.append("Tua")
            i+=1
        if usia >= 36 and usia <= 45:
            Parubaya += 1
            dataClassifyUsia.append("Paruhbaya")
            i+=1
        if usia >= 25 and usia <= 35:
            Muda += 1
            dataClassifyUsia.append("Muda")
            i+=1
    return dataClassifyUsia
# getClassifyUsia()
```

```python
# Klasifikasi Nilai Pelatihan Function
def getClassifyNilaiTes1(data_nilaites1):
    dataClassifyNilaiTes1 = []
    sangat_baik = 0
    baik = 0
    kurang = 0
    for nilaites1 in data_nilaites1:
        if nilaites1 >=81 and nilaites1 <= 100:
            sangat_baik += 1
            dataClassifyNilaiTes1.append("Sangat Baik")
        if nilaites1 >= 66 and nilaites1 <= 80:
            baik += 1
            dataClassifyNilaiTes1.append("Baik")
        if  nilaites1 >= 45 and nilaites1 <= 65:
            kurang += 1
            dataClassifyNilaiTes1.append("Kurang")
    return dataClassifyNilaiTes1
```

```python
# Klasifikasi Nilai Kinerja Function
def getClassifyNilaiTes2(data_nilaites2):
    dataClassifyNilaiTes2 = []
    sangat_baik = 0
    baik = 0
    kurang = 0
    for nilaites2 in data_nilaites2:
        if nilaites2 >=81 and nilaites2 <= 100:
            sangat_baik += 1
            dataClassifyNilaiTes2.append("Sangat Baik")
        if nilaites2 >= 66 and nilaites2 <= 80:
            baik += 1
            dataClassifyNilaiTes2.append("Baik")
        if  nilaites2 >= 45 and nilaites2 <= 65:
            kurang += 1
            dataClassifyNilaiTes2.append("Kurang")
    return dataClassifyNilaiTes2
```

```python
# Ciri Tiap Kelas FUNCTION
def CiriperKelas(ciri, kelas, valueciri, valuekelas):
    i = 0
    result = 0
    for x in ciri:
        if ciri[i] == valueciri and kelas[i] == valuekelas:
            result = result + 1
        i = i + 1
    return result
```

```python
# Probability Function (PH - PX)
def Prob(value):
    Prob = float("{:.8f}".format(value / Total_Data))
    return Prob
```

```python
# Cari P(X|H) Function
def PXH(cirikelas, totalkelas):
    pxh = cirikelas / totalkelas
    return pxh
```

```python
# cek PX Pada Ciri
def PXPengalamanKerja(PengalamanKerja):
    PX = 0
    if PengalamanKerja == "Junior":
        PX = PX_Junior
    else:
        if PengalamanKerja == "Supervisor":
            PX = PX_Supervisor
        else:
            if PengalamanKerja == "Senior":
                PX = PX_Senior
    return PX
def PXUsia(Usia):
    PX = 0
    if Usia == "Tua":
        PX = PX_Tua
    if Usia == "Paruhbaya":
        PX = PX_Paruhbaya
    if Usia == "Muda":
        PX = PX_Muda
    return PX
def PXNilaiTes1(NilaiTes1):
    PX = 0
    if NilaiTes1 == "Kurang":
        PX = PX_Kurang
    if NilaiTes1 == "Baik":
        PX = PX_Baik
    if NilaiTes1 == "Sangat Baik":
        PX = PX_SangatBaik
    return PX
def PXNilaiTes2(NilaiTes2):
    PX = 0
    if NilaiTes2 == "Kurang":
        PX = PX_Tes2_Kurang
    if NilaiTes2 == "Baik":
        PX = PX_Tes2_Baik
    if NilaiTes2 == "Sangat Baik":
        PX = PX_Tes2_SangatBaik
    return PX
```

```python
# Mencari PX Total
def PX_Total(PXPengalamanKerja,PXUsia,PXNilaiTes1,PXNilaiTes2):
    total = PXPengalamanKerja * PXUsia * PXNilaiTes1 * PXNilaiTes2
    return float("{:.15f}".format(total))
```

```python
# Mencari P(H|X) Function
def PHX(PXH, PH, PX):
    PHX = (PXH * PH) / PX
    return float("{:.8f}".format(PHX))
```

```python
# Mencari Hasil Evaluasi
def getHasilEvaluasi(PHX_DITERIMA, PHX_DIPERTIMBANGKAN, PHX_DITOLAK):
    Diterima = PHX_DITERIMA
    Dipertimbangkan = PHX_DIPERTIMBANGKAN
    Ditolak = PHX_DITOLAK
    HasilEvaluasi = ""
    if Diterima > Ditolak:
        if Diterima > Dipertimbangkan:
            HasilEvaluasi = "DITERIMA"
        else:
            HasilEvaluasi = "DIPERTIMBANGKAN"
    else:
        if Dipertimbangkan > Ditolak:
            HasilEvaluasi = "DIPERTIMBANGKAN"
        else:
            HasilEvaluasi = "DITOLAK"
    return HasilEvaluasi
```

```python
# Main Function
def NaiveBaiyes():
    HasilEvaluasi = []
    i = 0
    while i < len(dataLatihAfterClassify):
        # Inisiasi Baris 

        PengalamanKerja = dataLatihAfterClassify.loc[i]["PengalamanKerja"]
        Usia = dataLatihAfterClassify.loc[i]["Usia"]
        NilaiTes1 = dataLatihAfterClassify.loc[i]["NilaiTes1"]
        NilaiTes2 = dataLatihAfterClassify.loc[i]["NilaiTes2"]

        # Inisiasi PX Total_Data
        PX = PX_Total(PXPengalamanKerja(PengalamanKerja),
                            PXUsia(Usia),
                            PXNilaiTes1(NilaiTes1),
                            PXNilaiTes2(NilaiTes2))

        # Cek DITERIMA
        PXHPengalamanKerjaDITERIMA = CiriPengalamanKerjaDITERIMA(PengalamanKerja)
        PXHUsiaDITERIMA = CiriUsiaDITERIMA(Usia)
        PXHNilaiTes1DITERIMA = CiriNilaiTes1DITERIMA(NilaiTes1)
        PXHNilaiTes2DITERIMA = CiriNilaiTes2DITERIMA(NilaiTes2)
        PXHDITERIMA_Total = PXH_Total(PXHPengalamanKerjaDITERIMA , PXHUsiaDITERIMA , PXHNilaiTes1DITERIMA , PXHNilaiTes2DITERIMA)
        PH = PH_DITERIMA
        PHX_DITERIMA = PHX(PXHDITERIMA_Total, PH, PX)
        
        #Cek DIPERTIMBANGKAN
        PXHPengalamanKerjaDIPERTIMBANGKAN = CiriPengalamanKerjaDIPERTIMBANGKAN(PengalamanKerja)
        PXHUsiaDIPERTIMBANGKAN = CiriUsiaDIPERTIMBANGKAN(Usia)
        PXHNilaiTes1DIPERTIMBANGKAN = CiriNilaiTes1DIPERTIMBANGKAN(NilaiTes1)
        PXHNilaiTes2DIPERTIMBANGKAN = CiriNilaiTes2DIPERTIMBANGKAN(NilaiTes2)
        PXHDIPERTIMBANGKAN_Total = PXH_Total(PXHPengalamanKerjaDIPERTIMBANGKAN , PXHUsiaDIPERTIMBANGKAN , PXHNilaiTes1DIPERTIMBANGKAN , PXHNilaiTes2DIPERTIMBANGKAN)
        PH = PH_DIPERTIMBANGKAN
        PHX_DIPERTIMBANGKAN = PHX(PXHDIPERTIMBANGKAN_Total, PH, PX)
        
        # Cek DITOLAK
        PXHPengalamanKerjaDITOLAK = CiriPengalamanKerjaDITOLAK(PengalamanKerja)
        PXHUsiaDITOLAK = CiriUsiaDITOLAK(Usia)
        PXHNilaiTes1DITOLAK = CiriNilaiTes1DITOLAK(NilaiTes1)
        PXHNilaiTes2DITOLAK = CiriNilaiTes2DITOLAK(NilaiTes2)
        PXHDITOLAK_Total = PXH_Total(PXHPengalamanKerjaDITOLAK , PXHUsiaDITOLAK , PXHNilaiTes1DITOLAK , PXHNilaiTes2DITOLAK)
        PH = PH_DITOLAK
        PHX_DITOLAK = PHX(PXHDITOLAK_Total, PH, PX)
        # Save to HasilEvaluasi List
        HasilEvaluasi.append(getHasilEvaluasi(PHX_DITERIMA, PHX_DIPERTIMBANGKAN, PHX_DITOLAK))

        i = i + 1
    
    
    return HasilEvaluasi
```
